# Distribution Plots
Let's discuss some plots that allow us to vizaullisse the distribution of a data set. these plots are: 
* distplot 
* jointplot 
* pairplot 
* rugplot 
* kdeplot

# Imports

```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

# Data
seaborn comes with built in datasets

```python
tips = sns.load_dataset('tips')
```

```python
the distplot shows the distribution of a univariate set of obsservations
```

```python
sns.distplot(tips['total_bill'])
sns.despine(left=True)

# safe to ignore warning
```

```python
# to remove the kde layer(the curve) and just have the histogram 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(tips['total_bill'],edgecolor='black')
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
sns.despine(left=True)
```

```python
# joinplot
```

```python
alows you to basically match up 2 distplots for bivariate data
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```
