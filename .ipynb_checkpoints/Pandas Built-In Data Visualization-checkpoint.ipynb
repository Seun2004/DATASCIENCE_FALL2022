# Plot Types 
There are several plot types

# Area

```python
df2.plot.area(alpha=0.6) 

# the higher the alpha the more visible the graphs become
```

```python
# Barplots 
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
plt.style.use('fivethirtyeight')
```

```python
# scatterplots
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='rainbow')
```

```python
# use s to indicate size 
df1.plot.scatter(x='A',y='B',s=df1['C']*60,cmap='red')
```

```python
# use s to indicate size 
plt.style.use('dark_background')
df1.plot.scatter(x='A',y='B',c='C',s=df1['C']*60,cmap='inferno')
```

```python
# Boxplots
df2.plot.box()
```

# Hexagonal Bin Plot
Useful for Bivariate Data, alternative to scatterplot

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=20,cmap='Oranges')
```
